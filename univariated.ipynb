Measures of Dispersion:
-Determine the variance, standard deviation, range, and interquartile range for product price
-How varied are the product prices? Are there any indicators of a significant spread in prices?


```python
import numpy as np
price_variance=np.var(df['price'])
price_std_dev=np.std(df['price'])
price_range=np.ptp(df['price'])
price_iqr=np.percentile(df['price'],75)-np.percentile(df['price'],25)
price_variance, price_std_dev, price_range, price_iqr
```

```python
#What's the average price point of products listed? How does this compare with the most common price point (mode)?
average_price = df['price'].mean()
mode_price = df['price'].mode()[0]
average_price, mode_price
```

 Unpacking Product Ratings

```python
reviews_variance=np.var(df['reviews'])
reviews_std_dev=np.std(df['reviews'])
reviews_range=np.ptp(df['reviews'])
reviews_iqr=np.percentile(df['reviews'],75)-np.percentile(df['reviews'],25)
reviews_variance, reviews_std_dev, reviews_range, reviews_iqr
```

```python
#How do customers generally rate products? Is there a common trend?
average_reviews = df['reviews'].mean()
mode_reviews = df['reviews'].mode()[0]
average_reviews, mode_reviews
```

```python
#Calculate the skewness and kurtosis for the reviews column.
reviews_skewness = df['reviews'].skew()
reviews_kurtosis = df['reviews'].kurtosis()
reviews_skewness, reviews_kurtosis
```

```python
#Are the reviews normally distributed, or do they lean towards higher or lower values?
plt.figure(figsize=(10, 6))
plt.hist(df['reviews'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Product Reviews on Amazon UK')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.show()
```
